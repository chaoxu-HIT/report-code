```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from sklearn.covariance import EmpiricalCovariance
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Ellipse
import matplotlib.patches as mpatches
from scipy.stats import gaussian_kde
from scipy.spatial.distance import cdist, mahalanobis
import os

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei']  # 或 'Arial Unicode MS' for Mac
plt.rcParams['axes.unicode_minus'] = False

# 定义4个优化合金的成分数据
optimized_alloys = {
    '#1': {'Nb': 100-12.26-21.35-1.98-1.96-0.51-4.34-4.35, 'Si': 12.26, 'Ti': 21.35, 
           'Al': 1.98, 'Cr': 1.96, 'Hf': 0.51, 'Zr': 4.34, 'V': 4.35},
    '#2': {'Nb': 100-15.06-26.16-1.97-1.98-0.71-4.64, 'Si': 15.06, 'Ti': 26.16, 
           'Al': 1.97, 'Cr': 1.98, 'Hf': 0.71, 'Zr': 4.64, 'V': 0},
    '#3': {'Nb': 100-14.51-25.72-1.98-1.87-0.52, 'Si': 14.51, 'Ti': 25.72, 
           'Al': 1.98, 'Cr': 1.87, 'Hf': 0.52, 'Zr': 0, 'V': 0},
    '#4': {'Nb': 100-12.59-35-1.77-2.58, 'Si': 12.59, 'Ti': 35, 
           'Al': 1.77, 'Cr': 2.58, 'Hf': 0, 'Zr': 0, 'V': 0}
}

def load_and_process_data(file_path, sheet_name=0):
    """加载并处理Excel数据"""
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    return df

def extract_composition_features(df, element_columns=None):
    """提取成分特征"""
    # 自动识别数据中的元素列
    # 常见的元素符号列表
    common_elements = ['Nb', 'Si', 'Ti', 'Al', 'Cr', 'Hf', 'Zr', 'V', 'Mo', 'W', 'Ta', 'B', 'Ge', 
                      'Fe', 'Sn', 'Co', 'Ni', 'Cu', 'Y', 'Re', 'Ru', 'C', 'N', 'O']
    
    # 从数据框中识别元素列
    element_cols = []
    for col in df.columns:
        if col in common_elements:
            element_cols.append(col)
    
    print(f"识别到的元素列: {element_cols}")
    
    # 提取成分数据
    composition_data = pd.DataFrame()
    for elem in element_cols:
        composition_data[elem] = df[elem].fillna(0)
    
    return composition_data, element_cols

def add_optimized_alloys(composition_data, element_cols):
    """添加优化合金数据"""
    # 创建优化合金的DataFrame
    opt_df = pd.DataFrame(optimized_alloys).T
    
    # 检查优化合金中是否有训练数据中不存在的元素
    missing_elements = []
    for elem in opt_df.columns:
        if elem not in element_cols and opt_df[elem].sum() > 0:
            missing_elements.append(elem)
    
    if missing_elements:
        print(f"警告: 优化合金包含训练数据中不存在的元素: {missing_elements}")
        print("这些元素将被忽略，可能影响t-SNE分析的准确性！")
    
    # 确保优化合金数据与训练数据有相同的列
    # 只保留训练数据中存在的元素
    for col in element_cols:
        if col not in opt_df.columns:
            opt_df[col] = 0
    
    # 只选择训练数据中存在的元素列
    opt_df = opt_df[element_cols]
    
    # 添加标签
    labels = ['Training'] * len(composition_data) + ['Optimized'] * len(opt_df)
    alloy_names = [''] * len(composition_data) + list(optimized_alloys.keys())
    
    # 合并数据
    all_data = pd.concat([composition_data, opt_df], ignore_index=True)
    
    print(f"数据维度: 训练数据 {len(composition_data)} 个样本, 优化合金 {len(opt_df)} 个样本")
    print(f"特征维度: {len(element_cols)} 个元素")
    
    return all_data, labels, alloy_names

def perform_tsne(data, perplexity=30, n_iter=1000, random_state=42):
    """执行t-SNE降维"""
    tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=n_iter, 
                random_state=random_state, init='pca', learning_rate='auto')
    tsne_results = tsne.fit_transform(data)
    return tsne_results


# ==================== 改进版距离度量分析函数 ====================

def compute_mahalanobis_distances(data, training_mask):
    """
    计算马氏距离 (Mahalanobis Distance) - 改进版
    
    ✅ 修复: 添加正则化以提高数值稳定性
    
    参数:
        data: 原始特征数据 (N x D)
        training_mask: 训练数据的布尔掩码
    
    返回:
        distances: 每个样本的马氏距离
        cov_matrix: 协方差矩阵
        mean_vector: 均值向量
    """
    training_data = data[training_mask]
    
    # 计算训练数据的均值和协方差矩阵
    mean_vector = np.mean(training_data, axis=0)
    cov_matrix = np.cov(training_data.T)
    
    # ✅ 修复3: 添加正则化以提高数值稳定性
    regularization = 1e-6
    cov_matrix_reg = cov_matrix + regularization * np.eye(len(cov_matrix))
    
    # 计算协方差矩阵的逆
    try:
        cov_inv = np.linalg.inv(cov_matrix_reg)
    except np.linalg.LinAlgError:
        # 如果即使正则化后仍奇异，使用伪逆
        print("警告: 即使正则化后协方差矩阵仍奇异，使用伪逆")
        cov_inv = np.linalg.pinv(cov_matrix)
    
    # 计算每个样本的马氏距离
    distances = np.zeros(len(data))
    for i in range(len(data)):
        diff = data[i] - mean_vector
        distances[i] = np.sqrt(diff.T @ cov_inv @ diff)
    
    return distances, cov_matrix, mean_vector


def compute_nearest_neighbor_distances(data, training_mask):
    """
    计算最近邻距离 (Nearest Neighbor Distance)
    
    参数:
        data: 原始特征数据 (N x D)
        training_mask: 训练数据的布尔掩码
    
    返回:
        nn_distances: 每个样本到最近训练样本的距离
        nn_indices: 最近邻的索引
    """
    training_data = data[training_mask]
    training_indices = np.where(training_mask)[0]
    
    nn_distances = np.zeros(len(data))
    nn_indices = np.zeros(len(data), dtype=int)
    
    for i in range(len(data)):
        if training_mask[i]:
            # 对于训练样本，找第二近的邻居（排除自己）
            distances = np.sqrt(np.sum((training_data - data[i])**2, axis=1))
            distances[np.where(training_indices == i)[0]] = np.inf
            nn_idx = np.argmin(distances)
            nn_distances[i] = distances[nn_idx]
            nn_indices[i] = training_indices[nn_idx]
        else:
            # 对于优化样本，找最近的训练样本
            distances = np.sqrt(np.sum((training_data - data[i])**2, axis=1))
            nn_idx = np.argmin(distances)
            nn_distances[i] = distances[nn_idx]
            nn_indices[i] = training_indices[nn_idx]
    
    return nn_distances, nn_indices


def compute_local_density(data, training_mask, radius=2.5):
    """
    计算局部密度 (Local Density) - 改进版
    
    ✅ 修复1: 将半径从1.5增大到2.5，更适合高维空间
    
    参数:
        data: 原始特征数据 (N x D)
        training_mask: 训练数据的布尔掩码
        radius: 局部半径（以标准差为单位）
    
    返回:
        densities: 每个样本的局部密度（邻居数量）
    """
    training_data = data[training_mask]
    
    # 标准化数据
    mean = np.mean(training_data, axis=0)
    std = np.std(training_data, axis=0)
    std[std == 0] = 1  # 避免除以零
    
    densities = np.zeros(len(data))
    
    for i in range(len(data)):
        # 计算到所有训练样本的标准化距离
        normalized_diff = (training_data - data[i]) / std
        distances = np.sqrt(np.sum(normalized_diff**2, axis=1))
        
        # 计数在半径内的样本
        densities[i] = np.sum(distances < radius)
    
    return densities


def analyze_distance_metrics(data, labels, alloy_names):
    """
    综合分析所有距离度量 - 改进版
    
    ✅ 修复2: 使用基于实际数据的动态阈值进行风险评估
    
    返回详细的统计信息和可视化数据
    """
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    print("\n" + "="*80)
    print("距离度量分析 (Distance Metrics Analysis) - 改进版")
    print("="*80)
    
    # 1. 计算马氏距离
    print("\n1. 马氏距离分析 (Mahalanobis Distance)")
    print("-" * 80)
    maha_distances, cov_matrix, mean_vector = compute_mahalanobis_distances(data.values, training_mask)
    
    # 训练数据统计
    train_maha = maha_distances[training_mask]
    maha_median = np.median(train_maha)
    maha_95th = np.percentile(train_maha, 95)
    
    print(f"训练数据马氏距离统计:")
    print(f"  范围: [{np.min(train_maha):.3f}, {np.max(train_maha):.3f}]")
    print(f"  中位数: {maha_median:.3f}")
    print(f"  95百分位: {maha_95th:.3f}")
    
    # 优化合金统计
    opt_indices = np.where(optimized_mask)[0]
    opt_names = [n for n, m in zip(alloy_names, optimized_mask) if m]
    
    print(f"\n优化合金马氏距离:")
    for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
        maha_dist = maha_distances[idx]
        percentile = (np.sum(train_maha < maha_dist) / len(train_maha)) * 100
        print(f"  合金 {name}: D_M = {maha_dist:.3f} (第{percentile:.1f}百分位)")
        
        # 判断位置
        if maha_dist > np.max(train_maha):
            status = "超出训练范围 ⚠️"
        elif maha_dist > maha_95th:
            status = "接近边界"
        else:
            status = "在训练范围内 ✓"
        print(f"           状态: {status}")
    
    # 2. 计算最近邻距离
    print("\n2. 最近邻距离分析 (Nearest Neighbor Distance)")
    print("-" * 80)
    nn_distances, nn_indices = compute_nearest_neighbor_distances(data.values, training_mask)
    
    # 训练数据统计
    train_nn = nn_distances[training_mask]
    nn_median = np.median(train_nn)
    
    print(f"训练数据最近邻距离统计:")
    print(f"  范围: [{np.min(train_nn):.3f}, {np.max(train_nn):.3f}]")
    print(f"  中位数: {nn_median:.3f}")
    
    # 优化合金统计
    print(f"\n优化合金最近邻距离:")
    for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
        nn_dist = nn_distances[idx]
        nn_idx = nn_indices[idx]
        percentile = (np.sum(train_nn < nn_dist) / len(train_nn)) * 100
        
        print(f"  合金 {name}:")
        print(f"    距离: d_NN = {nn_dist:.3f} (第{percentile:.1f}百分位)")
        print(f"    最近邻索引: Sample_{nn_idx+1}")
        
        # 显示最近邻的成分（如果可能）
        if nn_idx < len(data):
            print(f"    最近邻成分: {dict(zip(data.columns, data.iloc[nn_idx]))}")
    
    # 3. 计算局部密度（改进版，使用半径2.5）
    print("\n3. 局部密度分析 (Local Density) - 改进版")
    print("-" * 80)
    local_densities = compute_local_density(data.values, training_mask, radius=2.5)
    
    # 训练数据统计
    train_density = local_densities[training_mask]
    density_median = np.median(train_density)
    
    print(f"训练数据局部密度统计 (半径=2.5σ):")
    print(f"  范围: [{int(np.min(train_density))}, {int(np.max(train_density))}] 个邻居")
    print(f"  中位数: {int(density_median)} 个邻居")
    
    # 优化合金统计
    print(f"\n优化合金局部密度:")
    for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
        density = local_densities[idx]
        percentile = (np.sum(train_density < density) / len(train_density)) * 100
        
        print(f"  合金 {name}:")
        print(f"    邻居数量: {int(density)} (第{percentile:.1f}百分位)")
        
        # 评估局部支撑（基于实际数据）
        if density >= density_median:
            support = "充足 ✓"
        elif density >= max(5, density_median * 0.5):
            support = "中等"
        else:
            support = "稀疏 ⚠️"
        print(f"    局部支撑: {support}")
    
    # ✅ 修复2: 综合评估 - 使用基于实际数据的动态阈值
    print("\n4. 综合外推风险评估 (改进版 - 基于数据动态阈值)")
    print("-" * 80)
    print(f"\n动态风险阈值 (基于训练数据分布):")
    print(f"  马氏距离: 低风险 < {maha_median:.2f} (中位数), "
          f"中等风险 < {maha_95th:.2f} (95百分位), 高风险 ≥ {maha_95th:.2f}")
    print(f"  最近邻距离: 低风险 < {nn_median:.2f} (中位数), "
          f"中等风险 < {nn_median*1.5:.2f} (1.5×中位数), 高风险 ≥ {nn_median*1.5:.2f}")
    print(f"  局部密度: 低风险 ≥ {int(density_median)} (中位数), "
          f"中等风险 ≥ {max(5, int(density_median*0.5))}, "
          f"高风险 < {max(5, int(density_median*0.5))}")
    
    for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
        print(f"\n合金 {name} 综合评估:")
        
        maha = maha_distances[idx]
        nn_dist = nn_distances[idx]
        density = local_densities[idx]
        
        # 风险评分（0-5，5为最安全）
        risk_score = 0
        
        # ✅ 马氏距离评分 - 基于实际分布
        if maha < maha_median:
            risk_score += 2
            maha_risk = "低"
        elif maha < maha_95th:
            risk_score += 1
            maha_risk = "中等"
        else:
            maha_risk = "高"
        
        # ✅ 最近邻距离评分 - 基于实际分布
        if nn_dist < nn_median:
            risk_score += 2
            nn_risk = "低"
        elif nn_dist < nn_median * 1.5:
            risk_score += 1
            nn_risk = "中等"
        else:
            nn_risk = "高"
        
        # ✅ 局部密度评分 - 基于实际分布
        if density >= density_median:
            risk_score += 1
            density_risk = "低"
        elif density >= max(5, density_median * 0.5):
            density_risk = "中等"
        else:
            density_risk = "高"
        
        print(f"  马氏距离风险: {maha_risk} (D_M={maha:.3f}, 阈值: 低<{maha_median:.2f}, 中<{maha_95th:.2f})")
        print(f"  最近邻风险: {nn_risk} (d_NN={nn_dist:.3f}, 阈值: 低<{nn_median:.2f}, 中<{nn_median*1.5:.2f})")
        print(f"  局部密度风险: {density_risk} (n={int(density)}, 阈值: 低≥{int(density_median)}, 中≥{max(5, int(density_median*0.5))})")
        print(f"  综合风险评分: {risk_score}/5")
        
        # 推荐
        if risk_score >= 4:
            recommendation = "✓ 高置信度，预测可靠性较高"
        elif risk_score >= 2:
            recommendation = "⚠ 中等置信度，建议实验验证"
        else:
            recommendation = "⚠⚠ 低置信度，强烈建议实验验证"
        
        print(f"  建议: {recommendation}")
    
    print("\n" + "="*80)
    
    # 返回所有计算结果
    return {
        'mahalanobis': maha_distances,
        'nearest_neighbor': nn_distances,
        'nn_indices': nn_indices,
        'local_density': local_densities,
        'cov_matrix': cov_matrix,
        'mean_vector': mean_vector
    }


def plot_distance_analysis(data, labels, alloy_names, distance_metrics, save_path=None):
    """
    创建距离分析的综合可视化图表 (Figure 6) - 带90th和95th百分位
    
    4个子图:
    (a) 马氏距离分布
    (b) 最近邻距离分布
    (c) 局部密度分布
    (d) 综合散点图
    """
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    maha_dist = distance_metrics['mahalanobis']
    nn_dist = distance_metrics['nearest_neighbor']
    density = distance_metrics['local_density']
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    
    # (a) 马氏距离分布
    ax = axes[0, 0]
    
    # 计算统计值
    train_maha = maha_dist[training_mask]
    maha_median = np.median(train_maha)
    maha_90th = np.percentile(train_maha, 90)
    maha_95th = np.percentile(train_maha, 95)
    
    # 添加背景色区域
    ax.axvspan(0, maha_median, alpha=0.15, color='green', label='Low risk (< Median)')
    ax.axvspan(maha_median, maha_90th, alpha=0.15, color='yellow', label='Medium risk (Median-90th)')
    ax.axvspan(maha_90th, maha_95th, alpha=0.15, color='orange', label='High risk (90th-95th)')
    ax.axvspan(maha_95th, ax.get_xlim()[1] if ax.get_xlim()[1] > maha_95th else maha_95th*1.2, 
               alpha=0.15, color='red', label='Very high risk (> 95th)')
    
    # 训练数据直方图
    ax.hist(maha_dist[training_mask], bins=30, alpha=0.6, color='lightblue', 
            edgecolor='navy', label='Training data', zorder=5)
    
    # 优化合金标记
    opt_indices = np.where(optimized_mask)[0]
    opt_names = [n for n, m in zip(alloy_names, optimized_mask) if m]
    alloy_colors = ['red', 'green', 'orange', 'purple']
    
    for i, (idx, name, color) in enumerate(zip(opt_indices, opt_names, alloy_colors)):
        ax.axvline(maha_dist[idx], color=color, linestyle='--', linewidth=2, 
                   label=f'Alloy {name}', zorder=10)
    
    # 添加统计线
    ax.axvline(maha_median, color='blue', linestyle=':', 
               linewidth=2, alpha=0.7, label='Median', zorder=8)
    ax.axvline(maha_90th, color='darkorange', linestyle=':', 
               linewidth=2, alpha=0.7, label='90th percentile', zorder=8)
    ax.axvline(maha_95th, color='darkred', linestyle=':', 
               linewidth=2, alpha=0.7, label='95th percentile', zorder=8)
    
    ax.set_xlabel('Mahalanobis Distance (D$_M$)', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_title('(a) Mahalanobis Distance Distribution', fontsize=13, fontweight='bold')
    ax.legend(fontsize=7, loc='upper right', ncol=2)
    ax.grid(True, alpha=0.3, zorder=1)
    
    # (b) 最近邻距离分布
    ax = axes[0, 1]
    
    # 计算统计值
    train_nn = nn_dist[training_mask]
    nn_median = np.median(train_nn)
    nn_90th = np.percentile(train_nn, 90)
    nn_95th = np.percentile(train_nn, 95)
    
    # 添加背景色区域
    ax.axvspan(0, nn_median, alpha=0.15, color='green', label='Low risk (< Median)')
    ax.axvspan(nn_median, nn_90th, alpha=0.15, color='yellow', label='Medium risk (Median-90th)')
    ax.axvspan(nn_90th, nn_95th, alpha=0.15, color='orange', label='High risk (90th-95th)')
    ax.axvspan(nn_95th, ax.get_xlim()[1] if ax.get_xlim()[1] > nn_95th else nn_95th*1.2, 
               alpha=0.15, color='red', label='Very high risk (> 95th)')
    
    # 训练数据直方图
    ax.hist(nn_dist[training_mask], bins=30, alpha=0.6, color='lightgreen', 
            edgecolor='darkgreen', label='Training data', zorder=5)
    
    # 优化合金标记
    for i, (idx, name, color) in enumerate(zip(opt_indices, opt_names, alloy_colors)):
        ax.axvline(nn_dist[idx], color=color, linestyle='--', linewidth=2, 
                   label=f'Alloy {name}', zorder=10)
    
    # 添加统计线
    ax.axvline(nn_median, color='green', linestyle=':', 
               linewidth=2, alpha=0.7, label='Median', zorder=8)
    ax.axvline(nn_90th, color='darkorange', linestyle=':', 
               linewidth=2, alpha=0.7, label='90th percentile', zorder=8)
    ax.axvline(nn_95th, color='darkred', linestyle=':', 
               linewidth=2, alpha=0.7, label='95th percentile', zorder=8)
    
    ax.set_xlabel('Nearest Neighbor Distance (d$_{NN}$)', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_title('(b) Nearest Neighbor Distance Distribution', fontsize=13, fontweight='bold')
    ax.legend(fontsize=7, loc='upper right', ncol=2)
    ax.grid(True, alpha=0.3, zorder=1)
    
    # (c) 局部密度分布
    ax = axes[1, 0]
    
    # 计算统计值
    train_density = density[training_mask]
    density_median = np.median(train_density)
    density_90th = np.percentile(train_density, 90)
    density_95th = np.percentile(train_density, 95)
    
    # 添加背景色区域 (注意：密度越高越好，所以颜色顺序相反)
    ax.axvspan(0, density_median*0.5, alpha=0.15, color='red', label='Very low density (< 50% Median)')
    ax.axvspan(density_median*0.5, density_median, alpha=0.15, color='orange', label='Low density (50% Median-Median)')
    ax.axvspan(density_median, density_90th, alpha=0.15, color='yellow', label='Medium density (Median-90th)')
    ax.axvspan(density_90th, ax.get_xlim()[1] if ax.get_xlim()[1] > density_90th else density_90th*1.2, 
               alpha=0.15, color='green', label='High density (> 90th)')
    
    # 训练数据直方图
    ax.hist(density[training_mask], bins=20, alpha=0.6, color='lightcoral', 
            edgecolor='darkred', label='Training data', zorder=5)
    
    # 优化合金标记
    for i, (idx, name, color) in enumerate(zip(opt_indices, opt_names, alloy_colors)):
        ax.axvline(density[idx], color=color, linestyle='--', linewidth=2, 
                   label=f'Alloy {name}', zorder=10)
    
    # 添加统计线
    ax.axvline(density_median, color='red', linestyle=':', 
               linewidth=2, alpha=0.7, label='Median', zorder=8)
    ax.axvline(density_90th, color='darkorange', linestyle=':', 
               linewidth=2, alpha=0.7, label='90th percentile', zorder=8)
    ax.axvline(density_95th, color='darkred', linestyle=':', 
               linewidth=2, alpha=0.7, label='95th percentile', zorder=8)
    
    ax.set_xlabel('Local Density (Number of neighbors within radius=2.5σ)', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_title('(c) Local Density Distribution', fontsize=13, fontweight='bold')
    ax.legend(fontsize=7, loc='upper right', ncol=2)
    ax.grid(True, alpha=0.3, zorder=1)
    
    # (d) 综合散点图: 马氏距离 vs 局部密度
    ax = axes[1, 1]
    
    # 训练数据
    ax.scatter(maha_dist[training_mask], density[training_mask], 
               c='lightblue', alpha=0.6, s=50, edgecolors='navy', 
               linewidth=0.5, label='Training data')
    
    # 优化合金
    for i, (idx, name, color) in enumerate(zip(opt_indices, opt_names, alloy_colors)):
        ax.scatter(maha_dist[idx], density[idx], c=color, s=300, marker='*', 
                   edgecolors='black', linewidth=2, label=f'Alloy {name}', zorder=10)
        
        # 添加标签
        ax.annotate(name, (maha_dist[idx], density[idx]), 
                   xytext=(5, 5), textcoords='offset points', 
                   fontsize=11, fontweight='bold')
    
    # 添加参考区域（基于实际数据）
    ax.axvspan(0, maha_median, alpha=0.1, color='green', 
               label=f'Low risk (D$_M$ < {maha_median:.1f})')
    ax.axvspan(maha_median, maha_90th, alpha=0.1, color='yellow', 
               label=f'Medium risk ({maha_median:.1f}-{maha_90th:.1f})')
    ax.axvspan(maha_90th, maha_95th, alpha=0.1, color='orange', 
               label=f'High risk ({maha_90th:.1f}-{maha_95th:.1f})')
    ax.axvspan(maha_95th, ax.get_xlim()[1], alpha=0.1, color='red', 
               label=f'Very high risk (> {maha_95th:.1f})')
    
    ax.set_xlabel('Mahalanobis Distance (D$_M$)', fontsize=12)
    ax.set_ylabel('Local Density (neighbors)', fontsize=12)
    ax.set_title('(d) Mahalanobis Distance vs Local Density', fontsize=13, fontweight='bold')
    ax.legend(fontsize=7, loc='best', ncol=2)
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"距离分析图已保存: {save_path}")
    
    plt.show()
    
    return fig


# ==================== 原有函数保持不变 ====================

def plot_kde_heatmap(ax, points, extent, resolution=100):
    """在指定的axis上绘制KDE热图"""
    if len(points) < 2:
        return None
    
    # 计算KDE
    kde = gaussian_kde(points.T)
    
    # 创建网格
    x_min, x_max, y_min, y_max = extent
    xx, yy = np.mgrid[x_min:x_max:complex(0, resolution), 
                      y_min:y_max:complex(0, resolution)]
    positions = np.vstack([xx.ravel(), yy.ravel()])
    
    # 计算KDE值
    f = np.reshape(kde(positions).T, xx.shape)
    
    # 绘制热图
    im = ax.imshow(f.T, extent=extent, origin='lower', 
                   cmap='YlOrRd', alpha=0.5, aspect='auto')
    
    # 添加等高线（可选）
    ax.contour(xx, yy, f, colors='gray', alpha=0.3, linewidths=0.5)
    
    return im


def plot_tsne_results(tsne_results, labels, alloy_names, title, save_path=None):
    """绘制t-SNE结果（带KDE热图背景）"""
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # 分离训练数据和优化合金数据
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    # 获取训练数据点
    training_points = tsne_results[training_mask]
    
    # 计算绘图范围
    im = None
    if len(training_points) > 0:
        margin = 0.2
        x_range = tsne_results[:, 0].max() - tsne_results[:, 0].min()
        y_range = tsne_results[:, 1].max() - tsne_results[:, 1].min()
        extent = [tsne_results[:, 0].min() - margin * x_range,
                  tsne_results[:, 0].max() + margin * x_range,
                  tsne_results[:, 1].min() - margin * y_range,
                  tsne_results[:, 1].max() + margin * y_range]
        
        # 绘制KDE热图（作为背景）
        if len(training_points) > 10:  # 只有足够的点才绘制KDE
            im = plot_kde_heatmap(ax, training_points, extent)
    
    # 绘制训练数据
    ax.scatter(tsne_results[training_mask, 0], tsne_results[training_mask, 1], 
               c='lightblue', alpha=0.8, s=50, label='Training data', 
               edgecolors='navy', linewidth=0.5, zorder=5)
    
    # 绘制优化合金
    colors = ['red', 'green', 'orange', 'purple']
    for i, (idx, name) in enumerate(zip(np.where(optimized_mask)[0], 
                                       [n for n, m in zip(alloy_names, optimized_mask) if m])):
        ax.scatter(tsne_results[idx, 0], tsne_results[idx, 1], 
                  c=colors[i], s=200, marker='*', label=f'Alloy {name}', 
                  edgecolors='black', linewidth=2, zorder=10)
        
        # 添加标签
        ax.annotate(name, (tsne_results[idx, 0], tsne_results[idx, 1]), 
                   xytext=(5, 5), textcoords='offset points', fontsize=12, 
                   fontweight='bold', zorder=15)
    
    # 添加置信椭圆（95%置信区间）
    if len(training_points) > 2:
        cov = np.cov(training_points.T)
        mean = np.mean(training_points, axis=0)
        
        # 计算椭圆参数
        eigenvalues, eigenvectors = np.linalg.eig(cov)
        angle = np.degrees(np.arctan2(eigenvectors[1, 0], eigenvectors[0, 0]))
        width, height = 2 * np.sqrt(5.991 * eigenvalues)  # 95% confidence
        
        ellipse = Ellipse(mean, width, height, angle=angle, 
                         facecolor='none', edgecolor='blue', 
                         linestyle='--', linewidth=2, alpha=0.8, zorder=6)
        ax.add_patch(ellipse)
    
    ax.set_xlabel('t-SNE Component 1', fontsize=14)
    ax.set_ylabel('t-SNE Component 2', fontsize=14)
    ax.set_title(title, fontsize=16, fontweight='bold')
    ax.legend(loc='best', fontsize=10)
    ax.grid(True, alpha=0.3, zorder=1)
    
    # 添加色阶图例
    if im is not None:
        cbar = plt.colorbar(im, ax=ax, pad=0.02)
        cbar.set_label('Density', fontsize=12)
        cbar.ax.tick_params(labelsize=10)
    
    # 调整布局
    plt.tight_layout()
    
    # 打印坐标轴范围
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    print(f"\n{title} - 坐标轴范围:")
    print(f"  横坐标 (t-SNE Component 1): [{xlim[0]:.3f}, {xlim[1]:.3f}]")
    print(f"  纵坐标 (t-SNE Component 2): [{ylim[0]:.3f}, {ylim[1]:.3f}]")
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    plt.show()
    
    return fig, ax


def plot_tsne_on_axis(ax, tsne_results, labels, alloy_names, title):
    """在指定的axis上绘制t-SNE结果（带KDE热图背景）"""
    # 分离训练数据和优化合金数据
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    # 获取训练数据点
    training_points = tsne_results[training_mask]
    
    # 计算绘图范围
    im = None
    if len(training_points) > 0:
        margin = 0.2
        x_range = tsne_results[:, 0].max() - tsne_results[:, 0].min()
        y_range = tsne_results[:, 1].max() - tsne_results[:, 1].min()
        extent = [tsne_results[:, 0].min() - margin * x_range,
                  tsne_results[:, 0].max() + margin * x_range,
                  tsne_results[:, 1].min() - margin * y_range,
                  tsne_results[:, 1].max() + margin * y_range]
        
        # 绘制KDE热图（作为背景）
        if len(training_points) > 10:  # 只有足够的点才绘制KDE
            im = plot_kde_heatmap(ax, training_points, extent, resolution=80)
    
    # 绘制训练数据
    ax.scatter(tsne_results[training_mask, 0], tsne_results[training_mask, 1], 
               c='lightblue', alpha=0.8, s=50, label='Training data', 
               edgecolors='navy', linewidth=0.5, zorder=5)
    
    # 绘制优化合金
    colors = ['red', 'green', 'orange', 'purple']
    for i, (idx, name) in enumerate(zip(np.where(optimized_mask)[0], 
                                       [n for n, m in zip(alloy_names, optimized_mask) if m])):
        ax.scatter(tsne_results[idx, 0], tsne_results[idx, 1], 
                  c=colors[i], s=200, marker='*', label=f'Alloy {name}', 
                  edgecolors='black', linewidth=2, zorder=10)
        
        # 添加标签
        ax.annotate(name, (tsne_results[idx, 0], tsne_results[idx, 1]), 
                   xytext=(5, 5), textcoords='offset points', fontsize=10, 
                   fontweight='bold', zorder=15)
    
    # 添加置信椭圆（95%置信区间）
    if len(training_points) > 2:
        cov = np.cov(training_points.T)
        mean = np.mean(training_points, axis=0)
        
        # 计算椭圆参数
        eigenvalues, eigenvectors = np.linalg.eig(cov)
        angle = np.degrees(np.arctan2(eigenvectors[1, 0], eigenvectors[0, 0]))
        width, height = 2 * np.sqrt(5.991 * eigenvalues)  # 95% confidence
        
        ellipse = Ellipse(mean, width, height, angle=angle, 
                         facecolor='none', edgecolor='blue', 
                         linestyle='--', linewidth=2, alpha=0.8, zorder=6)
        ax.add_patch(ellipse)
    
    ax.set_xlabel('t-SNE Component 1', fontsize=12)
    ax.set_ylabel('t-SNE Component 2', fontsize=12)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.legend(loc='best', fontsize=8)
    ax.grid(True, alpha=0.3, zorder=1)
    
    # 返回坐标轴范围和图像对象（用于色阶图例）
    return ax.get_xlim(), ax.get_ylim(), im


def analyze_distances(tsne_results, labels, alloy_names):
    """分析优化合金到训练数据的距离（基于t-SNE空间）"""
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    training_points = tsne_results[training_mask]
    optimized_points = tsne_results[optimized_mask]
    optimized_names = [n for n, m in zip(alloy_names, optimized_mask) if m]
    
    # 计算每个优化合金到最近训练数据点的距离
    print("\n优化合金到训练数据的最小距离分析 (t-SNE空间):")
    print("-" * 50)
    
    for i, name in enumerate(optimized_names):
        distances = np.sqrt(np.sum((training_points - optimized_points[i])**2, axis=1))
        min_dist = np.min(distances)
        avg_dist = np.mean(distances)
        
        print(f"合金 {name}:")
        print(f"  - 到最近训练数据点的距离: {min_dist:.3f}")
        print(f"  - 到训练数据的平均距离: {avg_dist:.3f}")
        
        # 判断是否在训练数据的分布范围内
        training_center = np.mean(training_points, axis=0)
        training_radius = np.max(np.sqrt(np.sum((training_points - training_center)**2, axis=1)))
        opt_to_center = np.sqrt(np.sum((optimized_points[i] - training_center)**2))
        
        if opt_to_center > training_radius:
            print(f"  - 位置: 在训练数据分布范围之外 (探索新区域)")
        else:
            print(f"  - 位置: 在训练数据分布范围之内")
    
    print("-" * 50)


def save_distance_analysis_to_excel(data, labels, alloy_names, distance_metrics, save_path, dataset_name):
    """
    保存距离分析结果到Excel - 包含90th和95th百分位
    """
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    # 创建完整数据框
    df_all = pd.DataFrame({
        'Sample_ID': [f'Sample_{i+1}' if name == '' else name for i, name in enumerate(alloy_names)],
        'Type': labels,
        'Mahalanobis_Distance': distance_metrics['mahalanobis'],
        'Nearest_Neighbor_Distance': distance_metrics['nearest_neighbor'],
        'Nearest_Neighbor_Index': [f"Sample_{idx+1}" for idx in distance_metrics['nn_indices']],
        'Local_Density': distance_metrics['local_density']
    })
    
    # 添加百分位排名
    train_maha = distance_metrics['mahalanobis'][training_mask]
    train_nn = distance_metrics['nearest_neighbor'][training_mask]
    train_density = distance_metrics['local_density'][training_mask]
    
    df_all['Mahalanobis_Percentile'] = [
        (np.sum(train_maha < val) / len(train_maha)) * 100 
        for val in distance_metrics['mahalanobis']
    ]
    df_all['NN_Percentile'] = [
        (np.sum(train_nn < val) / len(train_nn)) * 100 
        for val in distance_metrics['nearest_neighbor']
    ]
    df_all['Density_Percentile'] = [
        (np.sum(train_density < val) / len(train_density)) * 100 
        for val in distance_metrics['local_density']
    ]
    
    # 创建统计汇总 - 包含90th百分位
    stats_data = []
    stats_data.append({
        'Metric': 'Mahalanobis Distance',
        'Training_Min': np.min(train_maha),
        'Training_Median': np.median(train_maha),
        'Training_90th': np.percentile(train_maha, 90),
        'Training_95th': np.percentile(train_maha, 95),
        'Training_Max': np.max(train_maha)
    })
    stats_data.append({
        'Metric': 'Nearest Neighbor Distance',
        'Training_Min': np.min(train_nn),
        'Training_Median': np.median(train_nn),
        'Training_90th': np.percentile(train_nn, 90),
        'Training_95th': np.percentile(train_nn, 95),
        'Training_Max': np.max(train_nn)
    })
    stats_data.append({
        'Metric': 'Local Density',
        'Training_Min': np.min(train_density),
        'Training_Median': np.median(train_density),
        'Training_90th': np.percentile(train_density, 90),
        'Training_95th': np.percentile(train_density, 95),
        'Training_Max': np.max(train_density)
    })
    
    df_stats = pd.DataFrame(stats_data)
    
    # 保存到Excel
    excel_path = os.path.join(save_path, f'Distance_Analysis_{dataset_name}.xlsx')
    with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
        df_all.to_excel(writer, sheet_name='All Samples', index=False)
        df_stats.to_excel(writer, sheet_name='Statistics', index=False)
        
        # 只保存优化合金的详细信息
        df_optimized = df_all[df_all['Type'] == 'Optimized'].copy()
        df_optimized.to_excel(writer, sheet_name='Optimized Alloys', index=False)
    
    print(f"距离分析数据已保存: {excel_path}")
    
    return df_all, df_stats


def save_visualization_data_for_origin(data, labels, alloy_names, distance_metrics, save_path, dataset_name):
    """
    保存用于Origin可视化的详细数据 - 包含90th和95th百分位
    
    为三种距离度量生成独立的、易于绘图的数据表
    """
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    train_maha = distance_metrics['mahalanobis'][training_mask]
    train_nn = distance_metrics['nearest_neighbor'][training_mask]
    train_density = distance_metrics['local_density'][training_mask]
    
    opt_indices = np.where(optimized_mask)[0]
    opt_names = [n for n, m in zip(alloy_names, optimized_mask) if m]
    
    # Excel文件路径
    excel_path = os.path.join(save_path, f'Visualization_Data_for_Origin_{dataset_name}.xlsx')
    
    with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
        
        # ========== Sheet 1: 马氏距离分布数据 ==========
        # 计算直方图数据
        maha_hist, maha_bins = np.histogram(train_maha, bins=30)
        maha_bin_centers = (maha_bins[:-1] + maha_bins[1:]) / 2
        
        df_maha = pd.DataFrame({
            'Bin_Center': maha_bin_centers,
            'Frequency': maha_hist
        })
        
        # 添加统计线数据 - 包含90th
        df_maha_stats = pd.DataFrame({
            'Statistic': ['Median', '90th Percentile', '95th Percentile', 'Min', 'Max'],
            'Value': [
                np.median(train_maha),
                np.percentile(train_maha, 90),
                np.percentile(train_maha, 95),
                np.min(train_maha),
                np.max(train_maha)
            ]
        })
        
        # 添加优化合金数据
        opt_maha_data = []
        for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
            opt_maha_data.append({
                'Alloy': name,
                'Mahalanobis_Distance': distance_metrics['mahalanobis'][idx],
                'Percentile': (np.sum(train_maha < distance_metrics['mahalanobis'][idx]) / len(train_maha)) * 100
            })
        df_maha_alloys = pd.DataFrame(opt_maha_data)
        
        # 保存到不同的区域
        df_maha.to_excel(writer, sheet_name='Mahalanobis_Distance', startrow=0, index=False)
        df_maha_stats.to_excel(writer, sheet_name='Mahalanobis_Distance', startrow=len(df_maha)+3, index=False)
        df_maha_alloys.to_excel(writer, sheet_name='Mahalanobis_Distance', startrow=len(df_maha)+len(df_maha_stats)+6, index=False)
        
        # ========== Sheet 2: 最近邻距离分布数据 ==========
        nn_hist, nn_bins = np.histogram(train_nn, bins=30)
        nn_bin_centers = (nn_bins[:-1] + nn_bins[1:]) / 2
        
        df_nn = pd.DataFrame({
            'Bin_Center': nn_bin_centers,
            'Frequency': nn_hist
        })
        
        # 添加统计线数据 - 包含90th
        df_nn_stats = pd.DataFrame({
            'Statistic': ['Median', '90th Percentile', '95th Percentile', 'Min', 'Max'],
            'Value': [
                np.median(train_nn),
                np.percentile(train_nn, 90),
                np.percentile(train_nn, 95),
                np.min(train_nn),
                np.max(train_nn)
            ]
        })
        
        opt_nn_data = []
        for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
            opt_nn_data.append({
                'Alloy': name,
                'Nearest_Neighbor_Distance': distance_metrics['nearest_neighbor'][idx],
                'Percentile': (np.sum(train_nn < distance_metrics['nearest_neighbor'][idx]) / len(train_nn)) * 100,
                'Nearest_Neighbor_Index': f"Sample_{distance_metrics['nn_indices'][idx]+1}"
            })
        df_nn_alloys = pd.DataFrame(opt_nn_data)
        
        df_nn.to_excel(writer, sheet_name='Nearest_Neighbor_Distance', startrow=0, index=False)
        df_nn_stats.to_excel(writer, sheet_name='Nearest_Neighbor_Distance', startrow=len(df_nn)+3, index=False)
        df_nn_alloys.to_excel(writer, sheet_name='Nearest_Neighbor_Distance', startrow=len(df_nn)+len(df_nn_stats)+6, index=False)
        
        # ========== Sheet 3: 局部密度分布数据 ==========
        density_hist, density_bins = np.histogram(train_density, bins=20)
        density_bin_centers = (density_bins[:-1] + density_bins[1:]) / 2
        
        df_density = pd.DataFrame({
            'Bin_Center': density_bin_centers,
            'Frequency': density_hist
        })
        
        # 添加统计线数据 - 包含90th
        df_density_stats = pd.DataFrame({
            'Statistic': ['Median', '90th Percentile', '95th Percentile', 'Min', 'Max'],
            'Value': [
                np.median(train_density),
                np.percentile(train_density, 90),
                np.percentile(train_density, 95),
                np.min(train_density),
                np.max(train_density)
            ]
        })
        
        opt_density_data = []
        for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
            opt_density_data.append({
                'Alloy': name,
                'Local_Density': int(distance_metrics['local_density'][idx]),
                'Percentile': (np.sum(train_density < distance_metrics['local_density'][idx]) / len(train_density)) * 100
            })
        df_density_alloys = pd.DataFrame(opt_density_data)
        
        df_density.to_excel(writer, sheet_name='Local_Density', startrow=0, index=False)
        df_density_stats.to_excel(writer, sheet_name='Local_Density', startrow=len(df_density)+3, index=False)
        df_density_alloys.to_excel(writer, sheet_name='Local_Density', startrow=len(df_density)+len(df_density_stats)+6, index=False)
        
        # ========== Sheet 4: 散点图数据 (马氏距离 vs 局部密度) ==========
        df_scatter_train = pd.DataFrame({
            'Sample_ID': [f'Sample_{i+1}' for i in range(len(train_maha))],
            'Mahalanobis_Distance': train_maha,
            'Local_Density': train_density
        })
        
        opt_scatter_data = []
        for i, (idx, name) in enumerate(zip(opt_indices, opt_names)):
            opt_scatter_data.append({
                'Alloy': name,
                'Mahalanobis_Distance': distance_metrics['mahalanobis'][idx],
                'Local_Density': int(distance_metrics['local_density'][idx])
            })
        df_scatter_opt = pd.DataFrame(opt_scatter_data)
        
        df_scatter_train.to_excel(writer, sheet_name='Scatter_Plot_Data', startrow=0, index=False)
        df_scatter_opt.to_excel(writer, sheet_name='Scatter_Plot_Data', startrow=len(df_scatter_train)+3, index=False)
        
        # ========== Sheet 5: 绘图说明 ==========
        instructions = pd.DataFrame({
            'Sheet_Name': [
                'Mahalanobis_Distance',
                'Nearest_Neighbor_Distance', 
                'Local_Density',
                'Scatter_Plot_Data'
            ],
            'Data_Description': [
                '马氏距离分布直方图数据 + 统计线(含90th/95th) + 优化合金位置',
                '最近邻距离分布直方图数据 + 统计线(含90th/95th) + 优化合金位置',
                '局部密度分布直方图数据 + 统计线(含90th/95th) + 优化合金位置',
                '散点图数据: 训练数据点 + 优化合金点'
            ],
            'Plot_Type': [
                '直方图(Histogram) + 垂直线(Vertical Lines) + 背景区域',
                '直方图(Histogram) + 垂直线(Vertical Lines) + 背景区域',
                '直方图(Histogram) + 垂直线(Vertical Lines) + 背景区域',
                '散点图(Scatter Plot) + 星形标记(Star Markers)'
            ],
            'X_Axis': [
                'Bin_Center (马氏距离)',
                'Bin_Center (最近邻距离)',
                'Bin_Center (邻居数量)',
                'Mahalanobis_Distance'
            ],
            'Y_Axis': [
                'Frequency (频数)',
                'Frequency (频数)',
                'Frequency (频数)',
                'Local_Density (邻居数量)'
            ]
        })
        
        instructions.to_excel(writer, sheet_name='README', index=False)
        
        # ========== Sheet 6: 颜色方案 ==========
        colors = pd.DataFrame({
            'Element': ['Training Data', 'Alloy #1', 'Alloy #2', 'Alloy #3', 'Alloy #4', 
                       'Median Line', '90th Percentile Line', '95th Percentile Line'],
            'Suggested_Color': ['Light Blue', 'Red', 'Green', 'Orange', 'Purple', 
                               'Blue', 'Dark Orange', 'Dark Red'],
            'RGB_Code': ['(173,216,230)', '(255,0,0)', '(0,255,0)', '(255,165,0)', 
                        '(128,0,128)', '(0,0,255)', '(255,140,0)', '(139,0,0)'],
            'Line_Style': ['Bar', 'Dash', 'Dash', 'Dash', 'Dash', 'Dot', 'Dot', 'Dot']
        })
        
        colors.to_excel(writer, sheet_name='Color_Scheme', index=False)
    
    print(f"Origin可视化数据已保存: {excel_path}")
    print(f"  - 包含4个数据表 + 绘图说明 + 颜色方案")
    print(f"  - 统计数据包含90th和95th百分位")
    print(f"  - 可直接导入Origin进行专业绘图")
    
    return excel_path


def save_tsne_data(tsne_results, labels, alloy_names, element_cols, save_path, dataset_name):
    """保存t-SNE结果数据到Excel文件"""
    # 创建数据框
    df_results = pd.DataFrame({
        'Sample_ID': [f'Sample_{i+1}' if name == '' else name for i, name in enumerate(alloy_names)],
        'Type': labels,
        't-SNE_Component_1': tsne_results[:, 0],
        't-SNE_Component_2': tsne_results[:, 1]
    })
    
    # 分离训练数据和优化合金数据
    training_mask = np.array(labels) == 'Training'
    optimized_mask = np.array(labels) == 'Optimized'
    
    # 计算一些统计信息
    stats_data = []
    optimized_indices = np.where(optimized_mask)[0]
    optimized_names = [n for n, m in zip(alloy_names, optimized_mask) if m]
    
    for i, name in enumerate(optimized_names):
        opt_point = tsne_results[optimized_indices[i]]
        training_points = tsne_results[training_mask]
        
        # 计算距离
        distances = np.sqrt(np.sum((training_points - opt_point)**2, axis=1))
        min_dist = np.min(distances)
        avg_dist = np.mean(distances)
        
        # 判断是否在分布范围内
        training_center = np.mean(training_points, axis=0)
        training_radius = np.max(np.sqrt(np.sum((training_points - training_center)**2, axis=1)))
        opt_to_center = np.sqrt(np.sum((opt_point - training_center)**2))
        
        stats_data.append({
            'Alloy': name,
            'Min_Distance_to_Training': min_dist,
            'Avg_Distance_to_Training': avg_dist,
            'Distance_to_Center': opt_to_center,
            'Training_Radius': training_radius,
            'Within_Training_Distribution': 'Yes' if opt_to_center <= training_radius else 'No'
        })
    
    df_stats = pd.DataFrame(stats_data)
    
    # 保存到Excel
    excel_path = os.path.join(save_path, f't-SNE_results_{dataset_name}.xlsx')
    with pd.ExcelWriter(excel_path, engine='openpyxl') as writer:
        df_results.to_excel(writer, sheet_name='t-SNE Coordinates', index=False)
        df_stats.to_excel(writer, sheet_name='Distance Analysis', index=False)
        
        # 添加元素信息
        df_elements = pd.DataFrame({'Elements': element_cols})
        df_elements.to_excel(writer, sheet_name='Element Info', index=False)
    
    print(f"t-SNE数据已保存到: {excel_path}")
    
    return df_results, df_stats


# 主程序
def main():
    # 文件路径
    kq_file = r'D:\文成数据库\Nb-Si数据库6.21.2-成分-性能.xlsx'
    htc_file = r'D:\文成数据库\Nb-Si数据库2.28-高温压缩.xlsx'
    save_dir = r'D:\博一下\manuscript3-机器学习\nc\nc修改9.27\10.15\10.21\t-sne2'
    
    # 创建保存目录
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
        print(f"创建目录: {save_dir}")
    else:
        print(f"使用输出目录: {save_dir}")
    
    # 分析KQ数据
    print("=" * 60)
    print("分析KQ（断裂韧性）数据集")
    print("=" * 60)
    
    try:
        # 加载KQ数据
        kq_df = load_and_process_data(kq_file)
        print(f"KQ数据集包含 {len(kq_df)} 个样本")
        
        # 提取成分特征
        kq_composition, kq_elements = extract_composition_features(kq_df)
        
        # 添加优化合金
        kq_all_data, kq_labels, kq_names = add_optimized_alloys(kq_composition, kq_elements)
        
        # 执行t-SNE
        print("执行t-SNE分析...")
        kq_tsne = perform_tsne(kq_all_data)
        
        # 绘制t-SNE结果
        kq_save_path = os.path.join(save_dir, 't-SNE_KQ_plot.png')
        plot_tsne_results(kq_tsne, kq_labels, kq_names, 
                         't-SNE Analysis of Nb-Si Alloys (KQ Dataset)', 
                         kq_save_path)
        
        # 保存t-SNE数据到Excel
        save_tsne_data(kq_tsne, kq_labels, kq_names, kq_elements, save_dir, 'KQ')
        
        # 分析t-SNE空间距离
        analyze_distances(kq_tsne, kq_labels, kq_names)
        
        # 距离度量分析
        print("\n" + "="*60)
        print("执行高级距离度量分析（改进版）...")
        print("="*60)
        
        # 计算所有距离度量（基于原始特征空间）
        kq_distance_metrics = analyze_distance_metrics(kq_all_data, kq_labels, kq_names)
        
        # 绘制距离分析图
        distance_plot_path = os.path.join(save_dir, 'Distance_Analysis_KQ.png')
        plot_distance_analysis(kq_all_data, kq_labels, kq_names, 
                             kq_distance_metrics, distance_plot_path)
        
        # 保存距离分析数据到Excel
        save_distance_analysis_to_excel(kq_all_data, kq_labels, kq_names, 
                                       kq_distance_metrics, save_dir, 'KQ')
        
        # ========== 新增：保存Origin可视化数据 ==========
        print("\n保存Origin可视化数据...")
        save_visualization_data_for_origin(kq_all_data, kq_labels, kq_names,
                                          kq_distance_metrics, save_dir, 'KQ')
        
    except Exception as e:
        print(f"处理KQ数据时出错: {e}")
        import traceback
        traceback.print_exc()
    
    # 分析HTC数据
    print("\n" + "=" * 60)
    print("分析HTC（高温压缩）数据集")
    print("=" * 60)
    
    try:
        # 加载HTC数据
        htc_df = load_and_process_data(htc_file)
        print(f"HTC数据集包含 {len(htc_df)} 个样本")
        
        # 提取成分特征
        htc_composition, htc_elements = extract_composition_features(htc_df)
        
        # 添加优化合金
        htc_all_data, htc_labels, htc_names = add_optimized_alloys(htc_composition, htc_elements)
        
        # 执行t-SNE
        print("执行t-SNE分析...")
        htc_tsne = perform_tsne(htc_all_data)
        
        # 绘制t-SNE结果
        htc_save_path = os.path.join(save_dir, 't-SNE_HTC_plot.png')
        plot_tsne_results(htc_tsne, htc_labels, htc_names, 
                         't-SNE Analysis of Nb-Si Alloys (HTC Dataset)', 
                         htc_save_path)
        
        # 保存t-SNE数据到Excel
        save_tsne_data(htc_tsne, htc_labels, htc_names, htc_elements, save_dir, 'HTC')
        
        # 分析t-SNE空间距离
        analyze_distances(htc_tsne, htc_labels, htc_names)
        
        # 距离度量分析
        print("\n" + "="*60)
        print("执行高级距离度量分析（改进版）...")
        print("="*60)
        
        # 计算所有距离度量（基于原始特征空间）
        htc_distance_metrics = analyze_distance_metrics(htc_all_data, htc_labels, htc_names)
        
        # 绘制距离分析图
        distance_plot_path = os.path.join(save_dir, 'Distance_Analysis_HTC.png')
        plot_distance_analysis(htc_all_data, htc_labels, htc_names, 
                             htc_distance_metrics, distance_plot_path)
        
        # 保存距离分析数据到Excel
        save_distance_analysis_to_excel(htc_all_data, htc_labels, htc_names, 
                                       htc_distance_metrics, save_dir, 'HTC')
        
        # ========== 新增：保存Origin可视化数据 ==========
        print("\n保存Origin可视化数据...")
        save_visualization_data_for_origin(htc_all_data, htc_labels, htc_names,
                                          htc_distance_metrics, save_dir, 'HTC')
        
    except Exception as e:
        print(f"处理HTC数据时出错: {e}")
        import traceback
        traceback.print_exc()
    
    # 创建汇总的Excel文件
    try:
        summary_path = os.path.join(save_dir, 't-SNE_Summary.xlsx')
        with pd.ExcelWriter(summary_path, engine='openpyxl') as writer:
            # 添加说明页
            info_data = {
                'Dataset': ['KQ', 'HTC'],
                'Total_Samples': [len(kq_df) if 'kq_df' in locals() else 0, 
                                 len(htc_df) if 'htc_df' in locals() else 0],
                'Elements': ['Nb, Si, Ti, Al, Cr, Hf, Zr, Mo, V, W, Ta, Sn',
                            'Nb, Si, Ti, Al, Cr, Hf, Zr, Mo, Fe, B, V'],
                'Optimized_Alloys': ['#1, #2, #3, #4', '#1, #2, #3, #4']
            }
            pd.DataFrame(info_data).to_excel(writer, sheet_name='Summary', index=False)
            
            # 添加优化合金成分
            opt_comp_df = pd.DataFrame(optimized_alloys).T
            opt_comp_df.to_excel(writer, sheet_name='Optimized Alloy Compositions')
        
        print(f"\n汇总文件已保存: {summary_path}")
    except:
        pass
    
    # 创建组合图（两个数据集对比）
    try:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
        
        # 重新绘制KQ结果到ax1
        kq_xlim, kq_ylim, kq_im = None, None, None
        if 'kq_tsne' in locals():
            kq_xlim, kq_ylim, kq_im = plot_tsne_on_axis(ax1, kq_tsne, kq_labels, kq_names, 
                            't-SNE Analysis of Nb-Si Alloys (KQ Dataset)')
        
        # 重新绘制HTC结果到ax2
        htc_xlim, htc_ylim, htc_im = None, None, None
        if 'htc_tsne' in locals():
            htc_xlim, htc_ylim, htc_im = plot_tsne_on_axis(ax2, htc_tsne, htc_labels, htc_names, 
                            't-SNE Analysis of Nb-Si Alloys (HTC Dataset)')
        
        # 为每个子图添加色阶图例
        if kq_im is not None:
            cbar1 = fig.colorbar(kq_im, ax=ax1, pad=0.02, fraction=0.046)
            cbar1.set_label('Density', fontsize=10)
            cbar1.ax.tick_params(labelsize=8)
        
        if htc_im is not None:
            cbar2 = fig.colorbar(htc_im, ax=ax2, pad=0.02, fraction=0.046)
            cbar2.set_label('Density', fontsize=10)
            cbar2.ax.tick_params(labelsize=8)
        
        plt.tight_layout()
        combined_path = os.path.join(save_dir, 't-SNE_Combined_plot.png')
        plt.savefig(combined_path, dpi=300, bbox_inches='tight')
        
        # 打印组合图的坐标轴范围
        print(f"\n组合图 - 坐标轴范围:")
        if kq_xlim and kq_ylim:
            print(f"  KQ数据集:")
            print(f"    横坐标 (t-SNE Component 1): [{kq_xlim[0]:.3f}, {kq_xlim[1]:.3f}]")
            print(f"    纵坐标 (t-SNE Component 2): [{kq_ylim[0]:.3f}, {kq_ylim[1]:.3f}]")
        if htc_xlim and htc_ylim:
            print(f"  HTC数据集:")
            print(f"    横坐标 (t-SNE Component 1): [{htc_xlim[0]:.3f}, {htc_xlim[1]:.3f}]")
            print(f"    纵坐标 (t-SNE Component 2): [{htc_ylim[0]:.3f}, {htc_ylim[1]:.3f}]")
        
        plt.close()
        print(f"组合图已保存到: {combined_path}")
    except:
        pass
    
    print("\n" + "=" * 80)
    print("分析完成总结")
    print("=" * 80)
    print("\n生成的文件包括：")
    print("\n1. t-SNE可视化:")
    print("   - t-SNE_KQ_plot.png")
    print("   - t-SNE_HTC_plot.png")
    print("   - t-SNE_Combined_plot.png")
    print("\n2. 距离度量分析图 (包含90th和95th百分位线和背景色):")
    print("   - Distance_Analysis_KQ.png (包含4个子图)")
    print("   - Distance_Analysis_HTC.png (包含4个子图)")
    print("\n3. Excel数据文件:")
    print("   - t-SNE_results_KQ.xlsx")
    print("   - t-SNE_results_HTC.xlsx")
    print("   - Distance_Analysis_KQ.xlsx (包含90th百分位)")
    print("   - Distance_Analysis_HTC.xlsx (包含90th百分位)")
    print("   - t-SNE_Summary.xlsx")
    print("\n4. Origin可视化数据文件:")
    print("   - Visualization_Data_for_Origin_KQ.xlsx (包含90th和95th百分位)")
    print("   - Visualization_Data_for_Origin_HTC.xlsx (包含90th和95th百分位)")
    print("   (包含: 直方图数据 + 统计线 + 散点图数据 + 绘图说明 + 颜色方案)")
    print("\n✅ 改进说明:")
    print("- 局部密度半径从1.5σ增大到2.5σ（更适合高维空间）")
    print("- 风险评分使用基于实际数据的动态阈值（更准确）")
    print("- 添加正则化处理协方差矩阵（提高数值稳定性）")
    print("- 所有子图添加了90th和95th百分位线 ⭐")
    print("- 所有子图添加了背景色区域显示风险等级 ⭐")
    print("- 所有Excel文件包含90th和95th百分位数据 ⭐")
    
    print(f"\n所有文件已保存到: {save_dir}")

if __name__ == "__main__":
    print("开始增强版t-SNE分析（包含高级距离度量 + 90th/95th百分位）✅")
    print("请确保已安装必要的库：")
    print("pip install pandas numpy scikit-learn matplotlib seaborn openpyxl scipy")
    print("-" * 80)
    main()
```
